# Линейные модели для регрессии
Линейные модели делают прогноз на основе линейной функции (linear finction), аргументами которой выступают признаки входных данных. Таким образом, мы предполагаем, что целевая переменная является линейной комбинацией признаков (на самом деле это, конечно же, не так)
$y=w[0]*x[0] + ... + w[p]*x[p] + b$  
Здесь:
- $x[0],...,x[p] $ - значения признаков входных данных (для некоторой точки данных);  
- $w[0], ..., w[p], b$ - параметры модели, которые в ходе обучения настраиваются (параметры наклона для каждого признака и параметр сдвига);  
- $y$ - целевой признак, который предсказывается  

Линейная модель хорошо подходит для наборов данных, когда количество признаков превышает количество точек данных.

## Обычная регрессия (метод наименьших квадратов)
**Ordinary least squares** - это наиболее традиционный метод регресси. При обучении параметры $w$ и $b$ подбираются таким образом, чтобы минимизировать *среднеквадратичную ошибку (mean squared error)*:
$I(w, b) = (y_{pred}[1] - y[1])^2 + ... + (y_{pred}[n]-y[n])^2  -> min$  

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
boston = load_boston()
X_train, X_test, y_train, y_test = train_test_split(boston.data,
                                                    boston.target,
                                                   random_state=3)
print('Размер тренировочного набора: ', X_train.shape)
print()
print(boston.DESCR)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)

# параметры наклона
print(f'Количество параметров в W: {len(linear_regression.coef_)}')
print(f'W = {linear_regression.coef_}')
print(f'b = {linear_regression.intercept_}')
print()

# качество модели
print(f'Коэффициент детерминации (R^2) на train = {linear_regression.score(X_train, y_train)}')
print(f'Коэффициент детерминации (R^2) на test = {linear_regression.score(X_test, y_test)}')
```

Проблема обычной линейной регрессии в том, что её сложность нельзя регулировать (нет вручную настраиваемых параметров)
## Гребневая регрессия
Формула гребневой регрессии аналогична той, что используется в обычном МНК, но отличие в том, что на параметры $W$ накладываются дополнительные ограничения: *все элементы $w$ должны быть близки к 0, то есть каждый признак должен иметь небольшое влияние на результат*   

Это ограничение является примером **регуляризации (regularization)** - явного ограничения модели для преодтвращения переобучения. 

Регуляризация в гребневой регрессии - это L2 регуляризация.

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge()
ridge_model.fit(X_train, y_train)

print(f'W = {ridge_model.coef_}')
print(f'b = {ridge_model.intercept_}')
print()

print('R^2 for train =', ridge_model.score(X_train, y_train))
print('R^2 for test =', ridge_model.score(X_test, y_test))
```

Параметр alpha в модели Ridge позволяет регулировать компромисс между простотой модели и качеством работы на обучающем наборе. **Увеличение alpha** заставляет коэффициенты $W$ уменьшаться (сжиматься к нулю). Это снижает качество работы модели на обучающем наборе, но может улучшить её обобщающую способность. **Уменьшение alpha** приводит к меньшему сжатию коэффициентов $W$. При близком к нулю alpha мы будем приближаться к простой линейной модели.

```python
ridge15_model = Ridge(alpha=15).fit(X_train, y_train)

print('R^2 for train =', ridge15_model.score(X_train, y_train))
print('R^2 for test =', ridge15_model.score(X_test, y_test))
print()

print(f'W = {ridge15_model.coef_}')
print(f'b = {ridge15_model.intercept_}')
```

```python
ridge0001_model = Ridge(alpha=0.0001).fit(X_train, y_train)

print('R^2 for train =', ridge0001_model.score(X_train, y_train))
print('R^2 for test =', ridge0001_model.score(X_test, y_test))
print()

print(f'W = {ridge0001_model.coef_}')
print(f'b = {ridge0001_model.intercept_}')
```

На графике можно видеть, как параметр aplha влияет на значения параметров $w[0], w[1], ..., w[p-1]$. Чем больше aplha, тем более значения $W$ сгруппированы около нуля.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(13,5))

plt.plot(ridge_model.coef_, 's', label='Параметры наклона при aplha=1')
plt.plot(ridge15_model.coef_, '^', label='Параметры наклона при aplha=15')
plt.plot(ridge0001_model.coef_, 'o', label='Параметры наклона при aplha=0.0001')

plt.xlabel('Индекс параметра в массиве W')
plt.ylabel('Оценка значения параметра')


plt.legend()
```

Стоит отметить, что при *достаточном объеме данных в обучающей выборке регуляризация становится менее важной*. Как можно заметить, особой разницы между Ridge и Linear regression мы не выявили. Но стоит уменьшить объем данных в обучающей выборке, и гребневая регрессия покажет лучший результат:

```python
linear_regression = LinearRegression().fit(X_train[:50], y_train[:50])
ridge_model = Ridge(alpha=1).fit(X_train[:50], y_train[:50])

print('R^2 for train (linear reg) = ', linear_regression.score(X_train[:50], y_train[:50]))
print('R^2 for train (ridge) = ', ridge_model.score(X_train[:50], y_train[:50]))
print('R^2 for test (linear reg) = ', linear_regression.score(X_test[:10], y_test[:10]))
print('R^2 for test (ridge) = ', ridge_model.score(X_test[:10], y_test[:10]))
```

Регуляризация становится менее важна (при возрастании объема данных), так как с возрастанием объема данных модели становится *сложнее* переобучиться

## Лассо
Как и гребневая регрессия, Лассо тоже сжимает коэффициенты до близких к нулю значений, но только с помощью L1 регуляризации


Результат L1 заключается в том, что некоторые коэффициенты становится *точно* равны нулю, то есть некоторые признаки исключаются из модели

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso().fit(X_train, y_train)

print(f'W = {lasso_model.coef_}')
print()

print('R^2 на train для Lasso: ', lasso_model.score(X_train, y_train))
print('R^2 на test для Lasso: ', lasso_model.score(X_test, y_test))
```

Lasso имеет параметр alpha для определения степени сжатия коэффициентов до нулевых значений (чем больше alpha, тем ближе к 0 -> тем проще модель). При этом есть параметр max_iter, позволяющий контролировать максимальное число итераций

```python
lasso001_model = Lasso(alpha = 0.01, max_iter=100000).fit(X_train, y_train)

print(f'W={lasso001_model.coef_}')
print()

print('R^2 на train', lasso001_model.score(X_train, y_train))
print('R^2 на test', lasso001_model.score(X_test, y_test))
```

Лассо можно применять в тех ситуациях, когда имеется большое число признаков и есть основания считать, что только часть из них важны. Lasso выбирает подмножество признаков и строит модель по ним.

## ElasticNet
Класс ElasticNet сочетает в себе L1 и L2 регуляризацию, поэтому на практике он работает лучше всего

```python
from sklearn.linear_model import ElasticNet
elastic_model = ElasticNet(alpha=0.0001, max_iter=100000).fit(X_train, y_train)

print(f'W = {elastic_model.coef_}')
print()

print('R^2 на train для ElasticNet =', elastic_model.score(X_train, y_train))
print('R^2 на test для ElasticNet =', elastic_model.score(X_test, y_test))
```

# Линейные модели для бинарной классификации
В случае бинарной классификации мы устаналиваем **пороговое значение** (чаще всего это 0) и делаем прогноз с помощью проверки неравенства:  
$ f = w[0]*x[0]+...+w[p]*x[p]+b > 0$  
Если неравенство выполняется, то прогнозируемый класс - положительный, в ином случае - отрицательный. Бинарный классификатор - это классификатор, который разделяет два класса с помощью линии, плоскости или гиперплоскости. **Прогноз для новой точки данных будет делаться на основании того, по какую сторону от разделяющей плоскости лежит эта новая точка**.

Обучение линейного бинарного классификатора (т.е. подбор параметров $W$ и $b$) классифицируется по следующим критериям:  
1) факт использования регуляризации и вид регуляризации  
2) измеряемые метрики качества

## Логистическая регрессия и метод опорных векторов
**logistic regression** (LR) и **support vector classifier** (SVC) -  модели для классификации. Обе модели имеют схожие *границы принятия решений* (*ground desicion boundary*). По умолчанию обе используют L2 регуляризацию, а степень регуляризации определяется параметром $C$ (чем меньше $C$, тем больше регуляризация: тем больше параметры w приближаются к нулю). Интересная деталь состоит в том, что при низком значении $C$ модель пытается подстроиться под "большинство" точек данных в обучающей выборке, а при высоком $C$ подчеркивается важность каждой отдельной точки данных.  


```python
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
print(breast_cancer.DESCR)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(breast_cancer.data,
                                                    breast_cancer.target,
                                                   random_state=5)
logreg = LogisticRegression(C=10, max_iter=10000).fit(X_train, y_train)
print('Точность Logistic Regression на train:', logreg.score(X_train, y_train))
print('Точность Logistic Regression на test:', logreg.score(X_test, y_test))

print('Истинные классы: ', y_test[:5])
print('Предсказанные классы: ', logreg.predict(X_test[:5]))
```

Посмотрим, какие признаки теряют важность при увеличении регуляризации (при уменьшении $C$)

```python
logreg100 = LogisticRegression(C=100, max_iter=10000).fit(X_train, y_train)
logreg5 = LogisticRegression(C=0.5, max_iter=10000).fit(X_train, y_train)
logreg001 = LogisticRegression(C=0.01, max_iter=10000).fit(X_train, y_train)

plt.figure(figsize=(15,7))

plt.plot(logreg100.coef_.T, 'o', label='C=100')
plt.plot(logreg5.coef_.T, 'v', label='C=0.5')
plt.plot(logreg001.coef_.T, '^', label='C=0.01')

plt.xticks(range(breast_cancer.data.shape[1]),
           breast_cancer.feature_names, rotation=90)


plt.xlabel('Признак')
plt.ylabel('Важность признака в разных моделях')
plt.ylim(-1,1)

plt.legend()
```

Чтобы получить более интерпретируемую модель, можно использовать L1 регуляризацию

```python
logreg = LogisticRegression(C=10, penalty='l1', solver='liblinear').fit(X_train, y_train)
print('Точность Logistic Regression с l1 регуляризацией на test: \n', logreg.score(X_test, y_test))
```

# Линейная мультиклассовая классификация
Многие линейные модели предназначены только для бинарной классификации (логистическая регрессия - исключение).  
**Подход "один против всех" (one-vs-rest)** - для каждого класса строится бинарная модель, которая пытается отделить  этот класс от остальных, в результате количество моделей равно количеству классов. Точка данных подается на вход каждому бинарному классификатору и *побеждает* тот класс, модель которого выдает наибольшее значение.  

```python
from sklearn.datasets import make_blobs
x, y = make_blobs()
plt.scatter(x[:, 0], x[:, 1])
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
```

```python
x.shape
```

```python
import numpy as np
from sklearn.svm import LinearSVC
linear_svm = LinearSVC(C=10, penalty='l2', max_iter=10000).fit(x[:80], y[:80])

print('Правильность SVM на новых данных:', linear_svm.score(x[80:], y[80:]))
print('Количество классов - ', len(np.unique(y)))
print('Размер W у построенной модели:', linear_svm.coef_.shape)
print('Размер b у построенной модели:', linear_svm.intercept_.shape)
```

Как можно заметить, в силу того, что применяется подход *one-vs-rest*, параметры модели представлены уже не в виде вектора и значения, а в виде матрицы и вектора (размерность определяется количеством классов)

```python
plt.scatter(x[:, 0], x[:, 1])
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')

line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b','r','g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.ylim(-10,15)
plt.xlim(-10,8)
plt.legend(['Класс 0', 'Класс 1', 'Класс 2'])
```

#  Резюме
 - Основной параметр линейных моделей - параметр *регуляризации* ($alpha$ в регрессии, $С$ в LinearSVC и LogisticRegression). Большое $alpha$ и маленькое $C$ означают бОльшую регуляризацию (простые модели)
 - Если вы предполагаете, что из всех признаков важны только некоторые, то лучше использовать l1 регуляризацию
 - Линейные модели быстро обучаются и быстро прогнозируют
 - Линейные модели хорошо работают с разреженными данными
 - Хорошая интерпретация предсказанной метки с помощью формул, но сложная интепретация того, как были получены обучаемые параметры
 - Линейные модели подходят для тех случаев, когда количество признаков превышает количество наблюдений
