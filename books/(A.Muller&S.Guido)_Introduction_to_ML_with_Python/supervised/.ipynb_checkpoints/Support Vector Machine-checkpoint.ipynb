# Ядерный метод опорных векторов
Ранее был рассмотрен линейный метод опорных векторов (LinearSVC). Ядерный метод опорных векторов (SVM) - это расширения рассмотренного линейного метода. Он позволяет строить более сложные модели, которые не сводятся к построению гиперплоскостей.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs(centers=4, random_state=4)
y = y%2

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('Линейно неразделимые классы')
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
```

Линейные модели в некоторым смысле ограничены, поскольку гиперплоскость не всегда обладает нужной "гибкостью". Один из способов преодолеть это - добавить новые **нелинейные** признаки. Тогда линейная модель станет "искусственно нелинейной".
