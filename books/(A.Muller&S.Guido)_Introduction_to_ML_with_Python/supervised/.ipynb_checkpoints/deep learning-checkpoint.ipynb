# Глубокое обучение
Нейронные сети имеют большие перспективы в разных сферах применения машинного обучения.


MLP - многослойные персептроны - можно рассматривать как обобщение линейный моделей.

целевой признак - это взвешеннная суммая входных признаков. "Взвешены" они по коэффициентам `w[0], w[1], ..., w[p]`

Нейронная сеть может содержать в себе скрытые слои (hidden layers). Чем больше скрытых слоев, тем сложнее сеть и тем, соответственно, сложнее модель. Скрытые слои принимают входные данные, преобразовывают их и передают дальше. 

На этом примере: входной слой передает входы скрытому слою (здесь он всего один). Для каждого элемента скрытого слоя выполняется вычисление взвешенной суммы признаков (причем веса для каждого элемента могут быть различны). Потом может быть применена *функция активации* (обычно это $tanh$ или $ReLU$). В итоге на выходе каждого элемента скрытого слоя получается какое-то число - это *выход нейрона скрытого слоя*. Они становятся входами следующего слоя (в данном случае выходного). Точно также вычисляется их взвешенная сумма (с новыми весами), а затем, возможно, применяется функция активации.


```python
from sklearn.datasets import fetch_california_housing, load_breast_cancer
from sklearn.model_selection import train_test_split
house = fetch_california_housing(as_frame=True)
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(house.data, house.target)
print(f'Размер тренировочной выборки - {X_train.shape}; размер тестовой выборки - {X_test.shape}')

display(house.frame)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor().fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
from sklearn.neural_network import MLPRegressor
neural_model = MLPRegressor(hidden_layer_sizes=(40),
                            random_state=19,
                            learning_rate='adaptive', verbose=True,
                            batch_size=150, activation='tanh')
```

Суть в том, что линейная модель строит линейную границу принятия решения. А нейронная сеть позволяет построить нелинейную, но относительно гладку границу принятия решения. Отметим, что саму границу формируют как раз скрытые нейроны: в зависимости от их количества будет "сглажена" граница (поскольку каждый отдельный нейрон вносит вклад в эту границу)

```python
neural_model.fit(X_train, y_train)
```

```python
print(f'Правильность на тестовом наборе: {neural_model.score(X_test, y_test)}')
```

К параметрам нейронной сети также относится параметр alpha, который отвечает за l2 регуляризацию (чем больше alpha, тем ближе коэф. `w[0], ...` к нулю).  
Важно помнить, что веса нейронной сети (все) задаются перед началом обучения случайным образом. Это значит, что в зависимости от стартового значения генератора псевдослучайных чисел модель может быть *разной*.


Чаще всего нейронные сети требуют того, чтобы входные признаки были одного масштаба (в идеале, чтобы мат.ожидание было = 0, а дисперсия = 1, то есть, что это было нормальное распределение для каждого признака).

```python
mean_train = X_train.mean(axis=0)
std_train = X_train.std(axis=0) # для каждого признака 

X_train_scaled = (X_train - mean_train) / std_train
X_test_scaled = (X_test - mean_train) / std_train
# то же самое преобразование для тестового набора

# вызов fit автоматически сбрасывает все, чему обучилась модель ранее
# поэтому можно и не создавать новую модель, а "переобучить" старую
neural_model=MLPRegressor(hidden_layer_sizes=(25, 20, 5), alpha=0.001, max_iter=500, random_state=5, learning_rate='adaptive')
neural_model.fit(X_train_scaled, y_train)
print('После стандартизации правильность на тестовом наборе: ', neural_model.score(X_test_scaled, y_test))
```

```python
from sklearn.neural_network import MLPClassifier
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target)

X_train_scaled = (X_train - X_train.mean(axis=0)) / X_train.std(axis=0)
X_test_scaled = (X_test - X_train.mean(axis=0)) / X_train.std(axis=0)

neural_model = MLPClassifier(hidden_layer_sizes=(29, 19, 7), max_iter=500)
neural_model.fit(X_train_scaled, y_train)

print('Правильность классификации на тестовом наборе (после нормализации): ', neural_model.score(X_test_scaled, y_test))
```

Sometimes looking at the learned coefficients of a neural network can provide insight into the learning behavior:  
- weights unstructured => not used at all
- if very large coefficients exist => need to change regularization (increase alpha)
...

```python
import matplotlib.pyplot as plt


fig, axes = plt.subplots(4, 4)
vmin, vmax = neural_model.coefs_[0].min(), neural_model.coefs_[0].max() 
for coef, ax in zip(neural_model.coefs_[0].T, axes.ravel()):
    ax.matshow(coef.reshape(3,10), cmap=plt.cm.gray, vmin=0.5*vmin, vmax=0.5*vmax)
    ax.set_xticks(())
    ax.set_yticks(())
plt.show()
```

```python
plt.figure(figsize=(25, 7))
plt.imshow(neural_model.coefs_[0], interpolation='none', cmap = 'viridis')
plt.yticks(range(30), cancer.feature_names)
plt.xlabel('Столбцы матрицы весов')
plt.ylabel('Признаки')
plt.colorbar()
print(f'Всего слоев -> {len(neural_model.coefs_)}')
print(f'Входных признаков (на {0}-м слое): {len(neural_model.coefs_[0])}')
for i in range(1, len(neural_model.coefs_)):
    print(f'Весов на {i}-м слое: {len(neural_model.coefs_[i])}')
```

Здесь темные цвета соответствуют отрицательным значениям, а светлые 0 положительным (**теплокарта**).

# Резюме
Нейронные сети способны обрабатывать информацию, содержающуюсь в больших объемах данных и строить очень сложные модели. Чаще всего они превосходят алгоритмы машинного обучения (за исключением тех случаев, когда можно обойтись простой моделью).  
Минусы нейронных сетей в том, что они требуют тщательной настройки параметров, долгих и мощных вычислений (и, соответственно, длительного времени обучения). Нейронные сети не требуют, но желают нормализованные данные (результат говорит за себя).

Настройка параметров нейронной сети - это отдельная наука.
Чаще всего идут от простого к сложному. Полезным показателем является количество весов (коэффициентов, которые "обучаются"). Если всего 100 признаков и используется сети с одним слоем из 100 нейронов, то всего будет $100*100=10000$ коэф. между входных и скрытым слоем, а также $100*1=100$ весов между скрытым и выходным слоем.

Иногда строят сложную модель в смысле количества весов (много нейронов или много слоев), а затем путем регуляризации и уменьшения нейронов модель упрощают (стараясь не ухудшить эффективность)
