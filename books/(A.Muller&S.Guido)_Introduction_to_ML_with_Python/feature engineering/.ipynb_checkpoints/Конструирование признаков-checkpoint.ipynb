# Типы данных и конструирование признаков

***Continuous features** - *непрерывные признаки* - это признаки, представленные числами (e.g. яркость пикселей, характеристики измерений и т.д.)  
**Categorical/Discrete features** - *категориальные/дискретные признаки* - это признаки, описываемые не числовыми значениями (e.g. бренд продукта, цвет, название улицы и т.д.)  

**feature engineering** - это процесс оптимальной подготовки данных для решения поставленной задачи.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv("C:/Users/tolik/information_technology/kaggle/house prices/data/train.csv", index_col='Id')
test = pd.read_csv("C:/Users/tolik/information_technology/kaggle/house prices/data/test.csv",  index_col='Id')
print(f'Всего записей в тренировочной: {len(train)}')
print(f'Всего записей в тестовой: {len(test)}')
display(train.head())
display(test.head())
```

Требуется предсказать цену для квартир по имеющимся признакам. Среди признаков есть как категориальные, так и непрерывные. Чтобы использовать методы машинного обучения, надо перевести категориальные признаки в числовые  

**Прямое кодирование (one-hot-encoding / one-out-of-N encoding)** - замена категориальной переменной одним или несколькими новыми признаками, которые принимают значение 0 или 1. Например, если категориальный признак имеет 4 категории, то можно создать 4 новых признака и для каждой точки данных поставить '1' в тот новый признак, который соответствует закодированной категории рассматриваемой точке данных. Для остальных признаков проставить '0'. Таким образом, для каждой точки данных только один из новых признаков будет равен 1.

Для начала нужно посмотреть все имеющееся данные, распределения значений. На этом этапе:

Чтобы выполнить прямое кодирование, можно использовать функцию из библиотеки pandas

```python
train_dummies = pd.get_dummies(train)
print('Размер после кодирования: ', train_dummies.shape)
print('Признаки после кодирования: ', list(train_dummies.columns))
train_dummies
```

```python
# Оставляем только признаки
features = list(train_dummies.columns)
features.remove('SalePrice')

X = train_dummies[features].values
y = train_dummies['SalePrice'].values

print(f'Форма массива X: {X.shape}')
print(f'Форма массива y: {y.shape}')
```
