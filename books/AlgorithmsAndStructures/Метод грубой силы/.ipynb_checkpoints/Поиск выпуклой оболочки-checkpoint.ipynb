 .   ,   $(x_1, y_1)$  $(x_2, y_2)$,   $a*x+b*y = c$,  $a = y_2-y_1, b = x_1-x_2, c=x_1*y_2-y_1*x_2$.      2 :       $a*x+b*y > c$,    - $a*x+b*y < c$.  ,          ,      $a*x+b*y - c$   

```python
def convex_full(S):
    n = len(S)
    result = []
    for i in range(n-1):
        p1 = S[i]
        for j in range(i+1, n):
            p2 = S[j]
            
            #  
            a = p2[1] - p1[1]
            b = p1[0] - p2[0]
            c = p1[0]*p2[1] - p1[1]*p2[0]
            f = lambda x, y: a*x + b*y - c
            
            #       
            k = 0
            x, y = S[k]
            while f(x, y) >= 0 and k < n-1:
                k+=1
                x, y = S[k]
            if k == n-1:
                result.append((S[i], S[j]))
            
            #       
            k = 0
            x, y = S[k]
            while f(x, y) <= 0 and k < n-1:
                k+=1
                x, y = S[k]
                
            if k == n-1:
                result.append((S[i], S[j]))
    return result
```

```python
Points = [(5, -1), (4, -2), (7, -2),
     (2, 2), (5, 1), (7, 1),
     (6, 2), (9, 2), (7, 3),
     (4, 4), (8, 4), (3, 1)]

convex = convex_full(Points)

for segment in convex:
    print(f'  {segment[0]}  {segment[1]}')
```
