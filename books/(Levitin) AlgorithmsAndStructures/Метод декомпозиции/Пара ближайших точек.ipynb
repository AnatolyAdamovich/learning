# Задача о паре ближайших точек
Пусть множество $S$ состоит из $(x_1, y_1), ..., (x_n, y_n)$ - точек на плоскости. Будем считать, что точки упорядочены по координате $x$. Тогда можно разделить $S$ на два подмножества $S_1, S_2$, состоящие из $n/2$ точек каждое (с помощью медианы среди $x_1, x_2, ... x_n$). Иными словами, если точки нанесены на плоскость, то можно провести вертикальную разделяющую линию, по разные стороны от которой будет одинаковое число точек.  

Далее можно рекурсивно найти пары ближайших точек в $S_1$ и $S_2$ и сравнить расстояния между ними: $d_1 - min_{S_1}, d_2 - min_{S_2}    =>  d = min(d_1, d_2)$

Но пара точек с минимальным расстоянием может лежать по разные стороны от разделяющей линии. Однако можно ограничиться рассмотрением всего лишь тех точек, которые попадают в *вертикальную полосу шириной* $2*d$):


Пусть $C_1$, $C_2$ - подмножества $S$, соответствующие левой и правой части вертикальной полосы
Теперь для каждой точки $P(x, y)$ из $C_1$ нужно рассмотреть точки в $C_2$ такие, что их ордината не выходит за пределы интервала $[y-d, y+d]$. При этом можно поддерживать $C_1$ и $C_2$ отсортированными в порядке возрастания координаты $y$.

```python
from math import sqrt, inf
import matplotlib.pyplot as plt
%matplotlib inline

# точка
class Point:
    def __init__(self, point_name, x, y):
        self.name = point_name
        self.x = x
        self.y = y
        
    def __repr__(self):
        return f'{self.name}({self.x},{self.y})'
    
# Евклидово расстояние
def dist(p1, p2):
    return sqrt((p1.x - p2.x)**2 + (p1.y-p2.y)**2) 

# Визуализация точек для грубого подхода
def draw_points(points, marks=None):
    xs = list(map(lambda p: p.x,
                 points))
    ys = list(map(lambda p: p.y,
                 points))
    colors = ['green'] * len(points)
    if marks is not None:
        point_1, point_2 = marks[0], marks[1]
        i, j = points.index(point_1), points.index(point_2)
        colors[i], colors[j] = "blue", "blue" 
    plt.scatter(xs, ys, color=colors)
    plt.show()
```

## Метод грубой силы

```python
def ClosestPointsBrute(points):
    n = len(points)
    min_dist = inf
    result = None
    for i in range(n-1):
        for j in range(i+1, n):
            if dist(points[i], points[j]) < min_dist:
                min_dist = dist(points[i], points[j])
                result = (min_dist, points[i], points[j])
    return result

S = [Point('A', 1, 5), Point('B', 3, -2), Point('C', 5, 5),
    Point('D', -2, 3), Point('E', 2, 2), Point('F', 0, 0),
    Point('G', 5, -4), Point('I', 0, 4), Point('H', 1.5, 2.9),
    Point('T', -1.3, 1.5), Point('S', 4, 0.1), Point('Y', 3.2, -1.2),
    Point('W', -1.2, -2.2), Point('R', -2, -3.5), Point('U', -0.5, -3),
    Point('K', 1.25, -1.6)]
draw_points(S)
```

```python
d, p1, p2 = ClosestPointsBrute(S)
print(f'Расстояние между {p1} и {p2} - {d}')
```

```python
draw_points(S, [p1, p2])
```

## Метод декомпозиции

```python
# вспомогательная функция с грубым подходом к поиску пары
def brute_calc(points):
    n = len(points)
    min_dist = inf
    result = None
    for i in range(n-1):
        p1 = points[i]
        for j in range(i+1, n):
            p2 = points[j]
            if dist(p1, p2) < min_dist:
                min_dist = dist(p1, p2)
                result = (min_dist, p1, p2)
    
    return result


# слияние двух подмножеств с сортировкой по координате y
def merge(points, left, right):
    mid = (left + right) // 2
    tmp = []
    i, j = left, mid
    while i < mid or j < right:
        if j == right or (i < mid and points[i].y < points[j].y):
            tmp.append(points[i])
            i += 1
        else:
            tmp.append(points[j])
            j += 1
    points[left:right] = tmp

    
    
def ClosestPoints(points):
    # сортируем по координате x
    points.sort(key=lambda p: p.x)
        
    # основная функция
    def closest_calc(points, l, r):        
        if (r-l) <= 3:
            ans = brute_calc(points[l:r])
            points[l:r] = sorted(points[l:r], key=lambda p: p.y)
            return ans
        else:
            # Нахождение пар в двух разных частях плоскости
            # две части, в которых точки отсортированы по y
            mid_index = (r+l) // 2
            mid_x = points[mid_index].x 
            d1 = closest_calc(points, l, mid_index)
            d2 = closest_calc(points, mid_index, r)
            ans = d2 if (d1[0] > d2[0]) else d1 
            
            # слияние двух частей (с сортировкой по y)
            merge(points, l, r)
            
            # поиск пар в полосе шириной 2d
            current_min_distance = ans[0] # здесь миним. d
            points_in_line_2d = list(filter(
                lambda p: abs(p.x - mid_x) < current_min_distance,
                points[l:r]))
            n = len(points_in_line_2d)
            for i in range(n-1):
                p1 = points_in_line_2d[i]
                for j in range(i+1, n):
                    p2 = points_in_line_2d[j]
                    if p1.y-p2.y > current_min_distance:
                        #  нет смысла проверять дальше,
                        # т.к. отсортировано по y
                        # значит все предыдущие точки 
                        # на ещё бОльшем расстоянии от i-той точки
                        # чем j-тая точка
                        break
                    ans = ans if (current_min_distance < dist(p1, p2)) else (dist(p1,p2), p1, p2)
            
            return ans
    
    return closest_calc(points, 0, len(points))
```

```python
S = [Point('A', 1, 5), Point('B', 3, -2), Point('C', 5, 5),
    Point('D', -2, 3), Point('E', 2, 2), Point('F', 0, 0),
    Point('G', 5, -4), Point('I', 0, 4), Point('H', 1.5, 2.9),
    Point('T', -1.3, 1.5), Point('S', 4, 0.1), Point('Y', 3.2, -1.2),
    Point('W', -1.2, -2.2), Point('R', -2, -3.5), Point('U', -0.5, -3),
    Point('K', 1.25, -1.6)]
```

```python
d, p1, p2 = ClosestPoints(S)
print(f'Расстояние между {p1} и {p2} - {d}')
```

```python
draw_points(S, [p1, p2])
```

Подход, основанный на методе декомпозиции, позволяет получить наивысший (на данный момент) класс эффективности для решения данной задачи - $\theta(n*log(n))$
