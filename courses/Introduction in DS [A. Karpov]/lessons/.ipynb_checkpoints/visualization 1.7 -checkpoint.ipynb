# Визуализация

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_about_students = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
display(data_about_students.shape)
data_about_students.head()
```

# Встроенные методы pandas
https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html

```python
data_about_students.gender.value_counts().plot.bar()
```

```python
data_about_students['math score'].plot.kde()
```

```python
data_about_students.plot.box()
```

# Seaborn - надстройка над matplotlib
Обладает чуть более простым синтаксисом и позволяет кастомизировать графики

```python
sns.scatterplot(x='reading score', y='math score', data=data_about_students)
```

```python
ax = sns.lmplot(x='reading score', y='math score', hue='gender',
           data=data_about_students)
ax.set_xlabels('reading')
ax.set_ylabels('math')
```

# Упражнения

```python
genome_data = pd.read_csv('data/genome_matrix.csv', index_col='Unnamed: 0')
graph = sns.heatmap(data=genome_data, cmap='viridis')
graph.xaxis.set_ticks_position('top')
graph.xaxis.set_tick_params(rotation=90)
```

```python
dota_data = pd.read_csv('data/dota_hero_stats.csv', index_col=0)
dota_data.head()
```

```python
def from_string_to_number(s):
    return len(s.split())
dota_data.n_of_roles = dota_data.roles.apply(from_string_to_number)
dota_data.n_of_roles.hist()
```

```python
iris_data = pd.read_csv('data/iris.csv', index_col=0)
iris_data.head()
```

```python
for column in iris_data:
    sns.displot(x=column, data=iris_data, kde='True')
```

```python
sns.violinplot(x='petal length', data=iris_data)
```

```python
cols = iris_data.columns[:-1].to_list()
sns.pairplot(iris_data, hue='species')
```
